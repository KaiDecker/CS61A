```python
from operator import add, sub
```

add为加法，sub为减法

定义f函数，当b小于零时用减法，大于时用加法

```python
def a_plus_abs_b(a, b):
    """Return a+abs(b), but without calling abs.

    >>> a_plus_abs_b(2, 3)
    5
    >>> a_plus_abs_b(2, -3)
    5
    >>> a_plus_abs_b(-1, 4)
    3
    >>> a_plus_abs_b(-1, -4)
    3
    """
    if b < 0:
        f = sub
    else:
        f = add
    return f(a, b)

print(a_plus_abs_b(2, 3))

print(a_plus_abs_b(2, -3))

print(a_plus_abs_b(-1, 4))

print(a_plus_abs_b(-1, -4))
```

想法是使用所有数的和减去最大数的平方，便得到另外两个较小数字的平方和

```python
def two_of_three(i, j, k):
    """Return m*m + n*n, where m and n are the two smallest members of the
    positive numbers i, j, and k.

    >>> two_of_three(1, 2, 3)
    5
    >>> two_of_three(5, 3, 1)
    10
    >>> two_of_three(10, 2, 8)
    68
    >>> two_of_three(5, 5, 5)
    50
    """
    return i * i + j * j + k * k - max(i, j, k) * max(i, j, k)

print(two_of_three(1, 2, 3))

print(two_of_three(5, 3, 1))

print(two_of_three(10, 2, 8))

print(two_of_three(5, 5, 5))
```

注意：最开始为n - 1，因为数n能被n整除，需要忽略掉

```python
def largest_factor(n):
    """Return the largest factor of n that is smaller than n.

    >>> largest_factor(15) # factors are 1, 3, 5
    5
    >>> largest_factor(80) # factors are 1, 2, 4, 5, 8, 10, 16, 20, 40
    40
    >>> largest_factor(13) # factors are 1, 13
    1
    """
    "*** YOUR CODE HERE ***"
    factor = n - 1
    while n % factor != 0:
        factor -= 1
    return factor

print(largest_factor(15))

print(largest_factor(80))

print(largest_factor(13))
```

length为步骤数量

```python
def hailstone(n):
    """Print the hailstone sequence starting at n and return its
    length.

    >>> a = hailstone(10)
    10
    5
    16
    8
    4
    2
    1
    >>> a
    7
    >>> b = hailstone(1)
    1
    >>> b
    1
    """
    "*** YOUR CODE HERE ***"
    length = 1
    print(n)
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = n * 3 + 1
        print(n)
        length += 1
    return length

print(hailstone(10))

print(hailstone(1))

print(hailstone(27))
```
